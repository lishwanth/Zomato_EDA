```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on = 'Country Code',how = 'left')
final_df.head()
```

### To check datatypes

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df['Country'].value_counts().index
```

```python
country_values = final_df['Country'].value_counts().values
```

### Top 3 Countries which using Zomato
    1.India - 94.39%
    2.USA - 4.73%
    3.UK - 0.87%

```python
plt.pie(country_values[:3], labels = country_names[:3],autopct = "%1.2f%%")# Check this
```

```python
final_df.columns
```

### Observation
1. When ratings are between 4.5 - 4.9 -> Excellent
2. When ratings are between 4.0 - 4.5 -> Very good
3. When ratings are between 3.5 - 3.9 -> Good
4. When ratings are between 3.0 - 3.4 -> Average
5. When ratings are between 2.5 - 2.9 -> Average
6. When ratings are between 2.0 - 2.4 -> Poor

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating count"})
ratings.head()
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating count", data = ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",hue = "Rating color" ,data = ratings,palette=["white","red","orange","yellow","green","green"])
```

### Observation
    1. Not rated count is very high
    2. Maximum number of ratings are between 2.5 - 4.4

#### Count Plot

```python
sns.countplot(x="Rating color",data = ratings,palette = ["white","red","orange","yellow","green","green"])
```
